Teniendo en cuenta que todas registran la experiencia previa de los funcionario lo más conveniente sería convertirlas en variables de tipo numérico y sumarlos para la creación de una nueva variable llamada MESES_EXPERIENCIA. Sin embargo, antes de realizar este procedimiento es conveniente realizar un análisis general del contenido de las mismas. 


```python
#Se define una lista con los nombres de las variables a inspeccionar
cols = [
    'MESES_EXPERIENCIA_PUBLICO',
    'MESES_EXP_NEG_PROPIO',
    'MESES_EXPERIENCIA_PRIVADO',
    'MESES_EXPERIENCIA_DOCENTE',
]
#Se define una función que reemplazan las comas por puntos y los valores nulos por ceros en la lista de columnas seleccionadas

def estandarizar_numericas(df, columnas):
    df = df.copy()
    for col in columnas:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(',', '.', regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df
```

```python
# Se imprime la información de las variables en la lista cols transformada en numérica
df = estandarizar_numericas(df, cols)
df.describe()
```

Al examinar la información estadística básicas de las cuatro columnas surgen muchas inquietudes:

- La variable MESES_EXPERIENCIA_PUBLICO presenta valores negativos y valores muy altos. Por ejemplo, mayores a 900 meses, equivalente a más de 75 años, los cuales evidentemente no son plausibles.
- La variable MESES_EXPERIENCIA_PRIVADO presenta una situación similar con valores mayores a los 900 meses. 
- La variable MESES_EXP_NEG_PROPIO contiene todas sus entradas iguales a 0, por lo que no ofrece información relevante para el estudio.
- La variable MESES_EXPERIENCIA_DOCENTE también presenta valores mayores a los 900 meses. 

Adicionalmente, la página del conjunto de datos proporciona muy poca documentación que respalde el dataframe. Los metadatos públicos en el portal de datos abiertos sólo incluyen la descripción general del conjunto y estadísticas de publicación, sin un diccionario de datos ni glosario. En particular, no se encontró un documento anexo ni comentario oficial que explique las consideraciones iniciales durante la recolección y publicación de los datos. La única referencia externa es la inclusión del conjunto en iniciativas anticorrupción (hoja de ruta nacional de datos abiertos y programa PIDA), lo que da contexto de política pero no aporta definiciones de campo. 

Por lo tanto, ante la falta de claridad sobre las inconsistencias y ante la falta de herramientas para resolver las mismas, descartaremos estas variables.

```python
#Revisamos las entradas de la variable asignación básica
df.groupby(['ASIG_BASICA']).count()
```

```python
#Convertimos la columna ASIG_BASICA a tipo numérico eliminando las comas y espacios en blanco
def estandarizar_montos_us(df, columnas):
    df = df.copy()
    for col in columnas:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(',', '', regex=False)  # elimina separador de miles
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df
df = estandarizar_montos_us(df, ['ASIG_BASICA'])
```

```python
#agrupamos por asignacion básica para verificar la conversión
df.groupby(['ASIG_BASICA']).count()
```

```python
#Busquemos algunas inconsistencias en la columna ASIG_BASICA
print(f'La variable contiene {df['ASIG_BASICA'].isnull().sum()} valores sin registros.')
print(f'La variable contiene {df[df['ASIG_BASICA']<=0].shape[0]} valores menores o iguales a cero.')
print(f'La variable contiene {df[df['ASIG_BASICA']>4e7].shape[0]} valores mayores a 40 millones.')
```

Los datos mayores a 40 millones nos resultan de especial interés pues en Colombia la máxima asignación básica mensual de un funcionario público es de alrededor de 30 a 40 millones. Así,  teniendo en cuenta que los 224 casos encontrados corresponden al 0.9% de los datos, eliminaremos estos registros.

```python
# Se eliminan los valores inconsistentes en la columna ASIG_BASICA

df = df[
    df['ASIG_BASICA'].notna() 
    & (df['ASIG_BASICA'] > 0)
    & (df['ASIG_BASICA'] <= 3e7)
].reset_index(drop=True)
df
```

```python
#La variable 'FECHA_VINCULACION' también es de tipo objeto, por lo que la cambiaremos a formado fecha
def estandarizar_fechas(df, columnas, formato=None):
    df = df.copy()
    for col in columnas:
        df[col] = pd.to_datetime(
            df[col],
            format=formato,
            errors='coerce'
        )
    return df

df = estandarizar_fechas(df, ['FECHA_VINCULACION'])
```

```python
df.info()
```

```python
# Agrupamos por fecha de vinculación para verificar la conversión y detectar valores atípicos. 
df.groupby(['FECHA_VINCULACION']).count()
```

```python
# Luego de hacer un análisis visual de los datos, encontramos que hay una fecha que claramente es un valor atípico y corresponde al 5 de abril de 2099. Procedemos a eliminar este valor del Dataframe.
df = df[df['FECHA_VINCULACION'] !='2099-04-05']
df.info()
```

```python
#Convertimos el resto de las columnas tipo objeto a tipo string
datos = df.select_dtypes(include='object').columns.tolist()
def estandarizar_texto(df, columnas):
    df = df.copy()
    for col in columnas:
        df[col] = (
            df[col]
            .astype("string")
            .str.strip()
            .str.upper()
        )
    return df

df = estandarizar_texto(df, datos)
```

```python
#Revisamos las columnas con entradas nulas
df.isnull().sum()
```

Note que hay 1690 entradas nulas en las variables NIVEL_ACADEMICO y NIVEL_FORMACION. que corresponden al 6.9% de los datos y 6 entradas nulas en la variable Fecha_Vinculación. En este caso, eliminaremos las entradas nulas de la fecha de vinculación y luego revisemos que tipo de registro hay en las variables NIVEL_ACADEMICO y NIVEL_FORMACIÓN

```python
df = df.dropna(subset=['FECHA_VINCULACION']).reset_index(drop=True)
df.info()
```

```python
#Agrupamos por nivel de formación para revisar sus entradas
df.groupby(['NIVEL_FORMACION']).count()
```

```python
#Verificamos los registros que tienen tanto nivel académico como nivel de formación nulos
df[df['NIVEL_ACADEMICO'].isna() & df['NIVEL_FORMACION'].isna()]
```

Podemos notar que las entradas nulas en ambas variables coinciden en los registros. Es decir, los funcionario para los que no se reportó nivel de formación tampoco se les reportó nivel académico. Adicionalmente, teniendo en cuenta la naturaleza de ambas, el contenido de la variable NIVEL_FORMACION puede ser redundante con la variable 'NIVEL_ACADEMICO', por lo que eliminaremos el nivel de formación.

Decidimos mantener los registros con nivel académico nulo. En este caso los valores nulos pueden representar a personas que, por motivos que desconocemos, no se les reportó el nivel de formación. Este hecho, resulta importante dentro de la naturaleza del estudio, pues no reportar la experiencia académica del funcionario puede poner en tela de juicio su trabajo en el rol que se le asigne, por lo que los datos se imputarán en la categoría 'SIN_INFORMACION'

```python
#Se elimina la columna NIVEL_FORMACION
df.drop(columns=['NIVEL_FORMACION'], inplace = True)
```

```python
#se llenan los valores nulos de NIVEL_ACADEMICO con 'SIN INFORMACION'
df['NIVEL_ACADEMICO'] = df['NIVEL_ACADEMICO'].fillna('SIN INFORMACION')
df.info()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 12))
# 1. Proporción de género por nivel académico
tabla = pd.crosstab(df['NIVEL_ACADEMICO'], df['SEXO'])
tabla_prop = tabla.div(tabla.sum(axis=1), axis=0)

tabla_prop.plot(
    kind='barh',
    stacked=True,
    ax=axes[0,0],
    color=['navy', 'gold'],
    alpha=0.7
)
axes[0, 0].set_title('Proporción de Género por Nivel Académico')
axes[0, 0].set_xlabel('Proporción')
axes[0, 0].set_ylabel('Nivel académico')
plt.legend(title='Género')
#2. Ingreso por nivel académico
sns.boxplot(
    data=df,
    x='NIVEL_ACADEMICO',
    y='ASIG_BASICA',
    ax=axes[0, 1],
    boxprops=dict(color='navy', alpha=0.7),
)
axes[0, 1].set_title('Ingreso por Nivel Académico')
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Ingreso por nivel académico y orden
sns.scatterplot(
    data=df,
    y='TIPO_NOMBRAMIENTO',
    x='ASIG_BASICA',
    hue='ORDEN',
    color=['navy', 'gold'],
    alpha=0.7,
    ax=axes[1, 0]
)
axes[1, 0].set_title('Distribución por Nivel Académico discriminado por orden de contratación')
axes[1, 0].tick_params(axis='x', rotation=45)

# 4. Ingreso vs edad
sns.scatterplot(
    data=df,
    x='FECHA_VINCULACION',
    y='ASIG_BASICA',
    ax=axes[1, 1],
    color='salmon'
)
axes[1, 1].set_title('Ingreso vs Fecha de vinculación')


# 5 tipo de nombramiento y nivel jerarquico
tabla = pd.crosstab(df['TIPO_NOMBRAMIENTO'], df['NIVEL_JERARQUICO_EMPLEO'])
tabla_prop = tabla.div(tabla.sum(axis=1), axis=0)

tabla_prop.plot(
    kind='barh',
    stacked=True,
    ax=axes[2,0],
    color=['navy', 'gold', 'salmon', 'olive', 'turquoise'],
    alpha=0.7
)
axes[2, 0].set_title('Proporción de nivel jerarquico por tipo de nombramiento')
axes[2, 0].set_xlabel('Proporción')
axes[2, 0].set_ylabel('tipo de nombramiento')


#6 clasificacion organica y nivel jerarquico
tabla = pd.crosstab(df['CLASIFICACION_ORGANICA'], df['NIVEL_JERARQUICO_EMPLEO'])
tabla_prop = tabla.div(tabla.sum(axis=1), axis=0)
tabla_prop.plot(
    kind='barh',
    stacked=True,
    ax=axes[2,1],
    color=['navy', 'gold', 'salmon', 'olive', 'turquoise'],
    alpha=0.7
)
axes[2, 1].set_title('Proporción de nivel jerarquico por naturaleza jurídica')
axes[2, 1].set_xlabel('Proporción')
axes[2, 1].set_ylabel('naturaleza jurídica')

plt.tight_layout()
plt.show()
```
